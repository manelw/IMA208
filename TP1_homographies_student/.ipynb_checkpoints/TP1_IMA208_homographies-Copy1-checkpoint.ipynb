The first step of the DLT algorithm is to normalize the image point.

**Question:** Recall why the normalization is important. Recall how to do it.

In the following, we will perform the normalization using the so called *homogeneous* coordinates of the keypoints. The main idea is to add a last coordinate equal to one for each point. The point (x,y) thus becomes (x,y,1) in homogeneous coordinate.

The advantage is that we will be able to do the normalization using a simple matrix product : X_norm = T@X.

**Fill** the blanks in the below implementation.

```python
def normalize_image_points(imageEucl):
    """
    Input: 2D list with x,y image points
    Output: normalized data points
    """

    print("Normalizing data using similarity matrix...")

    imageEucl = np.array(imageEucl)
    mean, std = np.mean(imageEucl, 0), np.std(imageEucl)
    #mean, std = ..., ... # To be filled, (elementwise) mean and std of the data points
    
    image = np.concatenate((imageEucl, np.ones((imageEucl.shape[0],1))),axis=1) # Image in homogeneous coordinates
    #image = ... # Image in homogeneous coordinate
    
    Transformation = np.array([[np.sqrt(2)/std,          0,                       0],
                               [0,                       np.sqrt(2)/std,          0],
                               [-np.sqrt(2)/std*mean[0], -np.sqrt(2)/std*mean[1], 1]])
    # Define a similarity matrix enabling to normalize the data. Write the cooefficients in Tranformation such that image_norm = image@Transformation
    #Transformation = np.array([[...,          0,                       0],
    #                           [0,                       ...,          0],
    #                           [..., ..., 1]])
    
    # Apply transformation on data points
    image_norm = image@Transformation
    
    # Retrieve normalized image in the original input shape (25, 2)
    image_norm = image_norm[:,0:2]

    return image_norm, Transformation.T


```

```python
def compute_matrix_A(points1, points2, no_points):
    """
    Input: Already normalized correspondences for image1 and image2
    Output: Matrix A as defined in Zisserman p. 91
    """

    A = []

    for i in range(0, no_points):
        x, y = points1[i, 0], points1[i, 1] # Denote the coordinate in image 1 x and y for shortcut
        x_prime, y_prime = points2[i, 0], points2[i, 1] # Denote the coordinate in image 2 x_prime and y_prime for shortcut

        # Create A_i according to the fomula given in class
        # here we are assuming w_i and w_i_prime are one
        A.append([0, 0, 0, -x, -y, -1, y_prime*x, y_prime*y, y_prime])
        A.append([x, y, 1, 0, 0, 0, -x_prime*x, -x_prime*y, -x_prime])
        #A.append([...])
        #A.append([...])
    print()
    print("Stacked matrix A shape:", np.shape(A))

    return np.asarray(A)

```

# Part 2: automatic homography estimation and panorama stitching

To bypass the flaws that we observe with the previous method, we now propose to use a robust homography estimation method, the RANSAC algorithm. As an application, we will consider panorama stitching: from two given overlapping images, create a single one which is a panorama. To do that, we will need to estimate the best homography to warp an image into the perspective of the other. In addition, to avoid the cumbersome choice of the matching points between the two images, we will do an automatic keypoint identification by using the SIFT features.

Let us first load the two considered images.

```python
img1c = cv2.cvtColor(cv2.imread('uttower_left.jpg'),cv2.COLOR_BGR2GRAY)
img2c = cv2.cvtColor(cv2.imread('uttower_right.jpg'),cv2.COLOR_BGR2GRAY)

img1 = img1c
img2 = img2c

plt.figure(),plt.imshow(img1,cmap='gray')
plt.figure(),plt.imshow(img2,cmap='gray')
```

To stitch the two images, we need to have access to pair of corresponding points between image 1 and image 2. To do that, we first need to extract some keypoints in both images (that will be matched later on). We will use here the SIFT descriptors.

**Question:** recall the main insights of the SIFT descriptors.

**Implement** the SIFT descriptors for both image 1 and image 2. You might be willing to use the cv2.SIFT_create and sift.detectAndCompute.

```python
sift = cv2.SIFT_create(contrastThreshold=0.15) # The larger the threshold, the smaller the number of points
kp1,desc1 = sift.detectAndCompute(img1, None) #kp1 = location of the keypoints, desc1 = corresponding descriptors.

#sift = ... # Creates an object of the SIFT class. The larger the threshold, the smaller the number of points
#kp1,desc1 = ... #kp1 = location of the keypoints, desc1 = corresponding descriptors.

# For visualization
img1c=cv2.drawKeypoints(img1,kp1,img1c,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img1c)
plt.show()

kp2,desc2 = sift.detectAndCompute(img2, None) #kp2 = location of the keypoints, desc2 = corresponding descriptors.
img2c=cv2.drawKeypoints(img2,kp2,img2c,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2c)
```

```python
#feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(desc1,desc2) 

#matches = bf.match(...,...) 
matches = sorted(matches, key = lambda x:x.distance)


# For visualizing the matches
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:100], img2, flags=2)
plt.figure(figsize=(20,12)),plt.imshow(img3),plt.show()
```

The following code enables to keep only the coordinates of the 100 best matches in the two images. We will use them as an input of a RANSAC, which will enable to do a robust homography estimation between the two images.

```python
D1 = np.ones((2,100)) # Coordinates in image 1 of the first best matching points
D2 = np.ones((2,100)) # Coordinates in image 2 of the first best matching points
for ii in range(100):
    D1[:,ii] = kp1[matches[ii].queryIdx].pt
    D2[:,ii] = kp2[matches[ii].trainIdx].pt

D1 = D1.astype(int)
D2 = D2.astype(int)
```

```python
#ORIGINAL
def ransac(D1, D2, threshold, iterations):
    D1 = D1.astype(float)
    D2 = D2.astype(float)
    
    
    D1_temp = np.ones((D1.shape[0]+1,D1.shape[1]))
    D1_temp[0:2,:] = D1
    D1 = D1_temp
    
    D2_temp = np.ones((D2.shape[0]+1,D2.shape[1]))
    D2_temp[0:2,:] = D2
    D2 = D2_temp    
    
    n = D1.shape[1] # Number of points in D1
    

    best_n_inlier = 0
    best_H = np.zeros((3, 3)) # Best homography matrix
    #best_H = ... # Best homography matrix. Fill the gap with a matrix of the right size full of 0.
    
    for i in range(iterations):
        s = np.random.choice(n, 4, replace=False) # Draw randomly 4 indices to extract 4 points among D1 and D2.
        # s = ... # Draw randomly 4 indices to extract 4 points among D1 and D2. You can use np.random.choice()
        
        x = D1[0, s] # Shortcut for the first coordinate of the points in D1
        y = D1[1, s] # Shortcut for the second coordinate of the points in D1
        xp = D2[0, s] # Shortcut for the first coordinate of the points in D2
        yp = D2[1, s] # # Shortcut for the second coordinate of the points in D2
        
        P = np.array([
            [x[0], y[0], 1, 0, 0, 0, -x[0]*xp[0], -y[0]*xp[0], -xp[0]],
            [0, 0, 0, x[0], y[0], 1, -x[0]*yp[0], -y[0]*yp[0], -yp[0]],
            [x[1], y[1], 1, 0, 0, 0, -x[1]*xp[1], -y[1]*xp[1], -xp[1]],
            [0, 0, 0, x[1], y[1], 1, -x[1]*yp[1], -y[1]*yp[1], -yp[1]],
            [x[2], y[2], 1, 0, 0, 0, -x[2]*xp[2], -y[2]*xp[2], -xp[2]],
            [0, 0, 0, x[2], y[2], 1, -x[2]*yp[2], -y[2]*yp[2], -yp[2]],
            [x[3], y[3], 1, 0, 0, 0, -x[3]*xp[3], -y[3]*xp[3], -xp[3]],
            [0, 0, 0, x[3], y[3], 1, -x[3]*yp[3], -y[3]*yp[3], -yp[3]],
        ])
        
        #P = np.array([
        #    [x[0], y[0], 1, 0, 0, 0, -x[0]*xp[0], -y[0]*xp[0], -xp[0]],
        #    [0, 0, 0, x[0], y[0], 1, -x[0]*yp[0], -y[0]*yp[0], -yp[0]],
        #    [x[1], y[1], 1, 0, 0, 0, -x[1]*xp[1], -y[1]*xp[1], -xp[1]],
        #    [0, 0, 0, x[1], y[1], 1, -x[1]*yp[1], -y[1]*yp[1], -yp[1]],
        #    [x[2], y[2], 1, 0, 0, 0, -x[2]*xp[2], -y[2]*xp[2], -xp[2]],
        #    [0, 0, 0, x[2], y[2], 1, -x[2]*yp[2], -y[2]*yp[2], -yp[2]],
        #    [x[3], y[3], 1, 0, 0, 0, -x[3]*xp[3], -y[3]*xp[3], -xp[3]],
        #    [0, 0, 0, x[3], y[3], 1, -x[3]*yp[3], -y[3]*yp[3], -yp[3]],
        #])
        
        _, _, V = np.linalg.svd(P, full_matrices=True)
        V = V.T # Permet d'avoir le meme resultat que matlab, potentiellement au chiffre pres
        H = np.reshape(V[:, -1], (3, 3)) # A verifier, initialement transpose => ne pas mettre car les resultats de reshape sont transposes entre matlab et python
        
        D2_new = np.dot(H, D1)
        
        D2_new[0, :] = np.divide(D2_new[0, :],D2_new[2, :])
        D2_new[1, :] = np.divide(D2_new[1, :],D2_new[2, :])
        D2_new[2, :] = np.divide(D2_new[2, :],D2_new[2, :])
        
        SD = np.sum((D2_new - D2) ** 2, axis=0)
        inliers = np.where(SD < threshold)[0]
        num_inlier = len(inliers)
        
        if num_inlier > best_n_inlier:
            best_H = H
            best_n_inlier = num_inlier
            point1 = D1[:2, inliers]
            point2 = D2[:2, inliers]
            in_res = SD[inliers]
    
    
    return best_H, best_n_inlier, point1, point2, in_res

# Example usage:
# H, num_inlier, point1, point2, in_res = ransac(D1, D2, threshold, num, iterations)
```
