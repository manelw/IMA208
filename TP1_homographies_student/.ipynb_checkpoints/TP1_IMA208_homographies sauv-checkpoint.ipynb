```python
def get_average_dist_to_origin(points):
    dist = (points - [0,0])**2
    dist = np.sum(dist, axis=1)
    dist = np.sqrt(dist)

    return np.mean(dist)
```

```python
def normalize_image_points(image):
    """
    Input: 2D list with x,y image points
    Output:
    """

    print()
    print("Normalizing data using similarity matrix...")

    image = np.array(image)
    mean, std = np.mean(image, 0), np.std(image)

    # define similarity transformation
    # no rotation, scaling using sdv and setting centroid as origin
    Transformation = np.array([[std/np.sqrt(2), 0, mean[0]],
                               [0, std/np.sqrt(2), mean[1]],
                               [0,   0, 1]])

    # apply transformation on data points
    Transformation = np.linalg.inv(Transformation)
    image = np.dot(Transformation, np.concatenate((image.T, np.ones((1, image.shape[0])))))

    # retrieve normalized image in the original input shape (25, 2)
    image = image[0:2].T

    print("translated origin:", np.mean(image, axis=0))
    print("average distance to origin:", get_average_dist_to_origin(image))

    return image, Transformation
```

```python
def compute_matrix_A(points1, points2, no_points):
    """
    Input: Normalized correspondences for image1 and image2
    Output: Matrix A as defined in Zisserman p. 91
    """

    A = []

    for i in range(0, no_points):
        x, y = points1[i, 0], points1[i, 1]
        x_prime, y_prime = points2[i, 0], points2[i, 1]

        # create A_i according to the eq. in the book
        # here we are assuming w_i is one
        A.append([0, 0, 0, -x, -y, -1, y_prime*x, y_prime*y, y_prime])
        A.append([x, y, 1, 0, 0, 0, -x_prime*x, -x_prime*y, -x_prime])

    print()
    print("Stacked matrix A shape:", np.shape(A))

    return np.asarray(A)

def compute_SVD(matrix_A):
    print()
    print("Computing SVD...")

    return np.linalg.svd(matrix_A)
```

```python
def get_vector_h(matrix_V):
    """
    Input: Matrix V from SVD of A
    Output: Unitary vector h (last column of V matrix of SVD)
    """
    print()
    print("Obtaining vector h...")

    h = matrix_V[-1,:] / matrix_V[-1,-1]

    return h
```

```python
def compute_homography(pts1,pts2):

    # set data points to numpy arrays
    image1 = np.array(pts1)
    image2 = np.array(pts2)
    no_points = image1.shape[0]

    # normalize data
    image1_normalized, T = normalize_image_points(image1)
    image2_normalized, T_prime = normalize_image_points(image2)

    # get matrix A for each normalized correspondence (dims 2*n x 9)
    A = compute_matrix_A(image1_normalized, image2_normalized, no_points)

    # compute SVD of A
    U, S, V = compute_SVD(A)

    # get last column of V and normalize it (this is the vector h)
    h = get_vector_h(V)

    # obtain homography (H tilde)
    print()
    print("Reshaping to get homography H_tilde...")
    H_tilde = h.reshape(3,3)

    # denormalize to obtain homography (H) using the transformations and generalized pseudo-inverse
    H = np.dot(np.dot(np.linalg.pinv(T_prime), H_tilde), T)

    print()
    print("Denormalized to obtain homography H for 2D data points...")
    print("Matrix H:")
    print(H)
    
    return H
```

```python
H = compute_homography(pts1,pts2)
homography_matrix_normalized = H / H[2, 2]
result = cv2.warpPerspective(img2, np.linalg.inv(homography_matrix_normalized),(3000,3000))
plt.imshow(result,cmap='gray')
```

# Part 2: automatic homography estimation and panorama stitching

```python
window_size = 25;
l = int(window_size/2);
```

```python
img1c = cv2.cvtColor(cv2.imread('uttower_left.jpg'),cv2.COLOR_BGR2GRAY)
img2c = cv2.cvtColor(cv2.imread('uttower_right.jpg'),cv2.COLOR_BGR2GRAY)

img1 = img1c
img2 = img2c

plt.figure(),plt.imshow(img1,cmap='gray')
plt.figure(),plt.imshow(img2,cmap='gray')
```

```python
sift = cv2.SIFT_create(contrastThreshold=0.15) # The larger the threshold, the smaller the number of points
kp1,desc1 = sift.detectAndCompute(img1, None) #kp1 = emplacement des keypoints, desc1 = descripteurs correspondants
img1c=cv2.drawKeypoints(img1,kp1,img1c,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img1c)
plt.show()

kp2,desc2 = sift.detectAndCompute(img2, None)
img2c=cv2.drawKeypoints(img2,kp2,img2c,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(img2c)
```

```python
#feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

matches = bf.match(desc1,desc2) # Faire coder match ?
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:100], img2, flags=2)
plt.figure(figsize=(20,12)),plt.imshow(img3),plt.show()
```

```python
D1 = np.ones((2,100))
D2 = np.ones((2,100))
for ii in range(100):
    D1[:,ii] = kp1[matches[ii].queryIdx].pt
    D2[:,ii] = kp2[matches[ii].trainIdx].pt

D1 = D1.astype(int)
D2 = D2.astype(int)
```

```python
def ransac(D1, D2, threshold, num, iterations):
    D1 = D1.astype(float)
    D2 = D2.astype(float)
    count = 0
    D1_temp = np.ones((D1.shape[0]+1,D1.shape[1]))
    D1_temp[0:2,:] = D1
    D1 = D1_temp

    
    D2_temp = np.ones((D2.shape[0]+1,D2.shape[1]))
    D2_temp[0:2,:] = D2
    D2 = D2_temp    
    
    #D1[2, :] = 1
    #D2[2, :] = 1
    
    n = D1.shape[1] # Number of points in D1
    best_n_inlier = 0
    best_H = np.zeros((3, 3)) # Best homography matrix
    
    for i in range(iterations):
        s = np.random.choice(n, num, replace=False)
        x = D1[0, s]
        y = D1[1, s]
        xp = D2[0, s]
        yp = D2[1, s]
        
        P = np.array([
            [x[0], y[0], 1, 0, 0, 0, -x[0]*xp[0], -y[0]*xp[0], -xp[0]],
            [0, 0, 0, x[0], y[0], 1, -x[0]*yp[0], -y[0]*yp[0], -yp[0]],
            [x[1], y[1], 1, 0, 0, 0, -x[1]*xp[1], -y[1]*xp[1], -xp[1]],
            [0, 0, 0, x[1], y[1], 1, -x[1]*yp[1], -y[1]*yp[1], -yp[1]],
            [x[2], y[2], 1, 0, 0, 0, -x[2]*xp[2], -y[2]*xp[2], -xp[2]],
            [0, 0, 0, x[2], y[2], 1, -x[2]*yp[2], -y[2]*yp[2], -yp[2]],
            [x[3], y[3], 1, 0, 0, 0, -x[3]*xp[3], -y[3]*xp[3], -xp[3]],
            [0, 0, 0, x[3], y[3], 1, -x[3]*yp[3], -y[3]*yp[3], -yp[3]],
        ])
        
        _, _, V = np.linalg.svd(P, full_matrices=True)
        V = V.T # Permet d'avoir le meme resultat que matlab, potentiellement au chiffre pres
        H = np.reshape(V[:, -1], (3, 3)) # A verifier, initialement transpose => ne pas mettre car les resultats de reshape sont transposes entre matlab et python
        
        D2_new = np.dot(H, D1)
        
        D2_new[0, :] = np.divide(D2_new[0, :],D2_new[2, :])
        D2_new[1, :] = np.divide(D2_new[1, :],D2_new[2, :])
        D2_new[2, :] = np.divide(D2_new[2, :],D2_new[2, :])
        
        SD = np.sum((D2_new - D2) ** 2, axis=0)
        inliers = np.where(SD < threshold)[0]
        num_inlier = len(inliers)
        
        if num_inlier > best_n_inlier:
            best_H = H
            best_n_inlier = num_inlier
            point1 = D1[:2, inliers]
            point2 = D2[:2, inliers]
            in_res = SD[inliers]
    
    
    return best_H, best_n_inlier, point1, point2, in_res

# Example usage:
# H, num_inlier, point1, point2, in_res = ransac(D1, D2, threshold, num, iterations)
```

```python
threshold = 1;
num = 4;
iterations = 1000;
[H, inliers, point1, point2, residue] = ransac(D1, D2, threshold,  num, iterations=iterations)
```
